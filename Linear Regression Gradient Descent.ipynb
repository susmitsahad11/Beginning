# Linear and Logistic Regression :

```python
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
random.seed(10)
X = np.random.rand(1000)*100
Y = 4*X + np.random.rand(1000)*5
plt.figure(figsize = (20,10))
plt.scatter(X,Y)
```

```python
def get_normal_solution(X,Y):
  A = np.concatenate([X.reshape(-1,1),np.ones(len(X)).reshape(-1,1)], axis=1)# pad extra feature to the existing features which has a constant value 1 (coefficient of this will be the intercept term)
  W = np.dot(np.dot(np.linalg.inv(np.dot(np.transpose(A),A)), np.transpose(A)), Y)
  return W
```

```python
get_normal_solution(X, Y)
```

```python
def gradient_descent_solution(X,Y):
  X_data = np.concatenate([np.ones(len(X)).reshape(-1,1),X.reshape(-1,1)], axis=1)
  #print(X_data.shape)
  alpha=0.0000000000005
  
  def derivative(theta):
    a = np.dot(np.transpose(X_data),X_data)
    #print(a.shape)
    b = np.dot(a,theta)
    #print(b.shape)
    c = np.dot(np.transpose(X_data),Y)
    #print(c.shape)
    z = b - c
    #print(z.shape)
    return z

  theta_1 = np.array([3,2])
  theta_2 = theta_1 - (alpha* derivative(theta_1))
  error = np.power(np.linalg.norm((theta_1 - theta_2), ord = 2),2)
  print(error)
  print(theta_2)
  print(derivative(theta_1))

  while  error > 0.01:
    theta_1 = theta_2
    theta_2 = theta_1 - (alpha* derivative(theta_1))
    error = np.power(np.linalg.norm(theta_1 - theta_2, ord = 2),2)
    
  return theta_1
```

```python
gradient_descent_solution(X,Y)
```
